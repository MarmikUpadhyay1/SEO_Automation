```python
import time
import random
import pyautogui
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

# Open log file in append mode
log_file = open('session_log.txt', 'a')

def log(message):
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    log_message = f"{timestamp} - {message}"
    print(log_message)
    log_file.write(log_message + "\n")

# Load configuration files (domains and keywords)
with open('domains.txt', 'r') as f:
    domains = [line.strip() for line in f]

with open('keywords.txt', 'r') as f:
    keywords = [line.strip() for line in f]

# Setup browser with Developer Tools and Device Mode activated
def setup_browser():
    chrome_options = Options()

    chrome_options.add_argument("--incognito")  # Incognito mode
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--user-agent=Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1")

    driver = webdriver.Chrome(options=chrome_options)
    driver.get("https://www.google.com")
    pyautogui.press('f12')
    time.sleep(3)
    pyautogui.hotkey('ctrl', 'shift', 'm')
    return driver

# Accept Google consent pop-up (if it appears)
def accept_consent(driver):
    try:
        consent_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'I agree')] | //button[contains(text(), 'Accept')] | //button[contains(text(), 'Agree')]"))
        )
        consent_button.click()
        log("Consent accepted.")
    except NoSuchElementException:
        log("Consent popup not found.")
    except TimeoutException:
        log("Timed out waiting for the consent popup.")
    except Exception as e:
        log("An error occurred while accepting consent: " + str(e))

def search_google(driver, keyword):
    try:
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys(keyword)
        search_box.send_keys(Keys.RETURN)
        time.sleep(3)  # Wait for the search results page to load
        log(f"Searched for keyword: {keyword}")
    except NoSuchElementException:
        log("Search box not found.")
    except Exception as e:
        log("An error occurred while searching: " + str(e))

def check_and_click_result(driver):
    try:
        results = driver.find_elements(By.XPATH, "//a[@href]")
        for result in results:
            url = result.get_attribute("href")
            if any(domain in url for domain in domains):
                log(f"Found matching domain: {url}. Visiting...")
                result.click()
                return True
        log("No matching domain found in the first 100 Google results.")
    except Exception as e:
        log("An error occurred while checking results: " + str(e))
    return False

# Visit a random domain if no match found in search results
def visit_random_domain(driver):
    random_domain = random.choice(domains)
    log(f"No matching domain found. Visiting {random_domain} directly.")
    driver.get(f"http://{random_domain}")

# Simulate scrolling and visiting behavior for the set duration
def simulate_scrolling(driver, duration):
    end_time = time.time() + duration
    log(f"Simulating scrolling for {duration} seconds.")
    while time.time() < end_time:
        driver.execute_script("window.scrollBy(0, document.body.scrollHeight / 3);")
        time.sleep(random.randint(5, 10))  # Random pauses during scrolling

# Main function to execute the browsing automation
def main():
    driver = setup_browser()

    try:
        # Accept cookies or consent
        accept_consent(driver)

        # Choose a random keyword and search on Google
        keyword = random.choice(keywords)
        log(f"Searching for: {keyword}")
        search_google(driver, keyword)

        # Check SERP for matching domains and visit if found
        if not check_and_click_result(driver):
            visit_random_domain(driver)

        # Simulate a realistic browsing session
        duration = random.randint(10, 20)  # Random session duration between 100 to 200 seconds
        simulate_scrolling(driver, duration)

    finally:
        # Ensure the browser closes after the session
        log("Session complete. Closing browser.")
        driver.quit()
        log_file.close()  # Close the log file

# Run the main function 10 times
if __name__ == "__main__": 
    main()
```

```python
import time
import random
import pyautogui
import asyncio
from selenium import webdriver
from playwright.async_api import async_playwright
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

log_file = open('session_log.txt', 'a')

def log(message):
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    log_message = f"{timestamp} - {message}"
    print(log_message)
    log_file.write(log_message + "\n")

# Load configuration files (domains and keywords)
def load_config():
    with open('domains.txt', 'r') as f:
        domains = [line.strip() for line in f]
    with open('keywords.txt', 'r') as f:
        keywords = [line.strip() for line in f]
    return domains, keywords

# Setup browser with realistic User-Agent and incognito mode
def setup_browser(headless=False):
    chrome_options = Options()
    chrome_options.add_argument("--incognito")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--start-maximized")
    driver = webdriver.Chrome(options=chrome_options)
    driver.get("https://www.google.com")

    # Open DevTools in device mode
    time.sleep(2)
    pyautogui.press('f12')
    time.sleep(2)
    pyautogui.hotkey('ctrl', 'shift', 'm')
    time.sleep(2)
    select_random_device()
    
    time.sleep(3)
    return driver

def select_random_device():
    base_x, base_y = 444, 188  # Replace with actual coordinates for the dropdown
    pyautogui.click(base_x, base_y)
    random_down_count = random.randint(2, 10)
    for _ in range(random_down_count):
        pyautogui.press('down')
        time.sleep(0.2)
    pyautogui.press('enter')

def accept_consent(driver):
    try:
        consent_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'I agree')] | //button[contains(text(), 'Accept')] | //button[contains(text(), 'Agree')]"))
        )
        consent_button.click()
        log("Consent accepted.")
    except (NoSuchElementException, TimeoutException):
        log("Consent popup not found or timed out.")

def search_google(driver, keyword):
    try:
        search_box = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "q")))
        search_box.send_keys(keyword)
        search_box.send_keys(Keys.RETURN)
        log(f"Searched for keyword: {keyword}")
    except NoSuchElementException:
        log("Search box not found.")

def check_and_click_result(driver, domains):
    try:
        results = WebDriverWait(driver, 10).until(lambda d: d.find_elements(By.XPATH, "//a[@href]"))
        for result in results:
            url = result.get_attribute("href")
            if url and any(domain in url for domain in domains):
                log(f"Found matching domain: {url}. Visiting...")
                result.click()
                return True
        log("No matching domain found in the first 100 Google results.")
    except Exception as e:
        log("An error occurred while checking results: " + str(e))
    return False

async def use_playwright_for_clicks(url):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=False)
        page = await browser.new_page()
        await page.goto(url)
        await page.keyboard.press("Tab")
        await page.keyboard.press("Enter")
        await asyncio.sleep(3)
        log(f"Interacted with {url} via Playwright")
        await browser.close()

def simulate_scrolling(driver, duration):
    end_time = time.time() + duration
    log(f"Simulating scrolling for {duration} seconds.")
    while time.time() < end_time:
        driver.execute_script("window.scrollBy(0, document.body.scrollHeight / 3);")
        time.sleep(random.uniform(2, 5))

async def main():
    domains, keywords = load_config()
    driver = setup_browser(headless=False)

    try:
        accept_consent(driver)

        keyword = random.choice(keywords)
        log(f"Searching for: {keyword}")
        search_google(driver, keyword)

        if not check_and_click_result(driver, domains):
            random_domain = random.choice(domains)
            await use_playwright_for_clicks(f"http://{random_domain}")

        duration = random.randint(100, 200)
        simulate_scrolling(driver, duration)

    finally:
        log("Session complete. Closing browser.")
        driver.quit()
        log_file.close()

# Run the main function with event loop handling
if __name__ == "__main__":
    try:
        # If running inside an existing event loop, use `await main()`
        if asyncio.get_event_loop().is_running():
            await main()
        else:
            asyncio.run(main())
    except RuntimeError:
        # For environments where `await` is required instead of `asyncio.run`
        asyncio.ensure_future(main())
```
